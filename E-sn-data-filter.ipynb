```python
# create dataframe from csv data file

import pandas as pd
from ast import literal_eval
df = pd.read_csv(my_path + 'KEYWORDS_FILTERED_PROCESSED_TEXT_CORPUS_v2.csv',usecols = ['post_id','month','year','group_name',
                                                                                       'keywords_matched','lemmatized_body'],
                 converters={'group_name': literal_eval})
df.info()
```
