```python
# create dataframe from csv data file

import pandas as pd
from ast import literal_eval
df = pd.read_csv(my_path + 'KEYWORDS_FILTERED_PROCESSED_TEXT_CORPUS_v2.csv',usecols = ['post_id','year','month','group_name',
                                                                                       'keywords_matched','lemmatized_body'])
df.info()
```

```python
df.head(5)
```

```python
df = df.explode('group_name')
df.reset_index(drop = True,inplace = True)
len(df)
```

```python
df.info()
```

```python
df.head(3)
```

```python
groups = ['sn','hanging','firearm','benzodiazepines','opioids',
          'antiemetic','barbiturates','acid regulator','cyanides','other preservatives',
          'plant-based poisons','household chemicals','ricin','nitric oxide']

for i in groups:
  df[i] = 0 # Create one column for each group

df.group_name = df.group_name.apply(literal_eval)

for i in range(len(df)):
  curr_lis = df.loc[i,'group_name']
  for name in curr_lis:
    df.loc[i,name] = df.loc[i,name] + 1
```

```python
df.info()
```

```python
%%time
import spacy
from spacy import displacy
nlp = spacy.load('en_core_web_lg')
```

```python
%%time
df['location_detected'] = ''
df['org_detected'] = ''
for i in range(len(df)):
  doc = nlp(df['lemmatized_body'][i])
  locations = []
  orgs = []
  if(len(doc.ents)>0):
    for ent in doc.ents:
      if ent.label_ in ['GPE']:
        locations.append(str(ent))
      if ent.label_ in ['ORG']:
        orgs.append(str(ent))
    if(len(locations)>0):
      df['location_detected'][i] = locations
    if(len(orgs)>0):
      df['org_detected'][i] = orgs
```

```python
df.info()
```

```python
import numpy as np
df['location_detected'].replace('', np.nan, inplace=True)
df['org_detected'].replace('', np.nan, inplace=True)
df.dropna(subset=['location_detected'], inplace=True)
df.dropna(subset=['org_detected'], inplace=True)
df.info()
```

```python
df.head(3)
```

```python
df.rename(columns={ df.columns[13]: 'acid_regulator'}, inplace=True)
df.rename(columns={ df.columns[15]: 'other_preservatives'}, inplace=True)
df.rename(columns={ df.columns[16]: 'plantbased_poisons'}, inplace=True)
df.rename(columns={ df.columns[17]: 'household_chemicals'}, inplace=True)
df.rename(columns={ df.columns[19]: 'nitric_oxide'}, inplace=True)
```

```python
df.to_csv(my_path+'NER.csv', sep=',', header=True, index=False)
```

```python
# save locations and organizations for manual verification

locations = df['location_detected'].tolist()
orgs = df['org_detected'].tolist()

import itertools
locations_list = list(itertools.chain(*locations))
orgs_list = list(itertools.chain(*orgs))

print(len(locations_list))
print(len(orgs_list))

from collections import Counter
locations_dict = Counter(locations_list)
orgs_dict = Counter(orgs_list)

with open(my_path+'organizations-lg.csv', 'w') as f:
  for org in orgs_dict:
    f.write(org)
    f.write(',')
    f.write(str(orgs_dict.get(org)))
    f.write('\n')

with open(my_path+'locations-lg.csv', 'w') as f:
  for loc in locations_dict:
    f.write(loc)
    f.write(',')
    f.write(str(locations_dict.get(loc)))
    f.write('\n')
```
